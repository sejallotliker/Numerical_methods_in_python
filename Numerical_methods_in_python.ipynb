#Integrals and derivatives in python:

##Trapezoidal Rule:
Suppose we have a function f(x) and we want to calculate its integral with respect to x from x = a to x = b

We can use:
1. Trapezoidal method

We divide the interval from a to b into N slices or steps

h = (b-a)/N

`A_k = 1/2* [ h*f(a+(k-1)h)+f(a+kh)]`




image.png

```python
#To apply trapezoidal rule:
def f(x):
  return x**4-2*x+1

N = 10
a= 0
b = 2.0
h = (b-a)/N

s = 0.5*(f(a)+f(b)) #initial value of s

for k in range(1, N): #Taking till N ensures that the for loop is impemented till N-1
  s += f(a+k*h)

print(s*h)
```

#Simpson's rule

###Simpson's rule involve fitting quadratic curves to piars of slices and then calculating the area under the quadratics

image.png

```python
#simpson's rule:

N=50
a=0
b=2
h=(b-a)/N

def f(x):
  return x**4-2*x+1

s= (f(a)+f(b))
for k in range(1, N, 2):
  s+=f(a+k*h)*4
for i in range(2, N, 2):
    s+=f(a+i*h)*2


print(h*s*(1/3))
```
